```python
def prep_fields(astr):
    return astr.replace('\n',' ').strip().split()

def summarize_values(headers):
    df = DATA[headers]
    dx =  df.describe(include='O').T
    #dx = dx.sort_values(by=['unique'], ascending=False )
    dx['rows_cnt'] = dx['count'].max()
    dx['nulls'] = dx['rows_cnt'] - dx['count']
    return dx

def sample_values(headers):
    return DATA[headers].sample(20).T

def count_values(headers, count=5):

    df = DATA[headers].copy()
    reader = df.iteritems()
    reader = ( pd.value_counts(ds, dropna=False).head(count) for name, ds in reader)
    df =pd.DataFrame(reader).stack().to_frame('values')
    return df
```
